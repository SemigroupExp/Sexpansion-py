# 8. Adjoint representation, Casimir operator and abelian subalgebras of $\mathfrak{sl}(2)$

Moving from semigroups to Lie algebras: this notebook exercises the structural tools the package
provides for a Lie algebra, using $\mathfrak{sl}(2, \mathbb{R})$ in the paper's basis

$$[X_1, X_2] = -2X_3, \qquad [X_1, X_3] = 2X_2, \qquad [X_2, X_3] = 2X_1.$$

It ports the Java example `II_AdjointRep_Casimirs_ex` from Section 5.5 of the paper (C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)).

```python
from sexpansion import LieAlgebra
from sexpansion.reports import adjoint_report, metric_report

sl2 = LieAlgebra.sl2()
```

## Adjoint representation

The adjoint matrices are built directly from the structure constants,
$(\mathrm{Adj}[X_i])^k{}_j = C_{ij}{}^k$. As everywhere in these notebooks, `one_based=True`
selects the paper's 1-based generator labels.

```python
print(adjoint_report(sl2, one_based=True))
```

## Cartan–Killing metric

The metric $g_{ij} = C_{ik}{}^l C_{jl}{}^k$ is the trace form of the adjoint representation. For
this basis it is $\mathrm{diag}(-8, 8, 8)$: non-degenerate, so the algebra is semi-simple, and
with signature $(+,-,0) = (2, 1, 0)$ it identifies the split real form
$\mathfrak{sl}(2, \mathbb{R})$ (the compact form $\mathfrak{su}(2)$ would be negative
definite).

```python
print(metric_report(sl2.cartan_killing_metric()))
```

## Quadratic Casimir operator

The quadratic Casimir in the adjoint representation is
$C = g^{ij}\, \mathrm{Adj}[X_i]\, \mathrm{Adj}[X_j]$ with $g^{ij}$ the inverse
Cartan–Killing metric. For a simple algebra, Schur's lemma forces it to be proportional to the
identity in any irreducible representation.

```python
print(metric_report(sl2.casimir(), precision=3))
print("Casimir eigenvalues:", sl2.casimir_eigenvalues())
```

## Maximal abelian subalgebra

A brute-force search for the largest set of mutually commuting generators, returned as 0-based
generator indices.

```python
print("Maximal abelian subalgebra (generator indices):", sl2.maximal_abelian_subalgebra())
```

## Result

The Casimir operator is exactly the identity (eigenvalues $1, 1, 1$), confirming that the adjoint
representation of $\mathfrak{sl}(2)$ is irreducible. The maximal abelian subalgebra is
one-dimensional — spanned by $X_1$ — matching the rank of $\mathfrak{sl}(2)$. This subalgebra is
precisely the grading subspace $V_0$ used in the resonant S-expansion of notebook 9.
