# 1. Checking associativity of multiplication tables

A finite semigroup is a set $S = \{\lambda_1, \dots, \lambda_n\}$ equipped with a closed,
**associative** product, usually presented as an $n \times n$ multiplication table. Closure is
guaranteed by construction, but associativity is not — so it is the first thing to verify before a
hand-entered table can be used in an S-expansion.

This notebook ports the Java example `I_associative_checking` from Section 5.1 of the paper
(C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). We enter three multiplication tables by hand — the two candidate tables of Figure 8 and
the semigroup $S_E^{(2)}$ of Figure 5 — and test each one.

```python
import numpy as np

from sexpansion import Semigroup
```

## The multiplication tables

Entry `table[i][j]` holds the index of the product $\lambda_i \cdot \lambda_j$. The package uses
**0-based element labels** internally, while the paper writes 1-based labels
$\lambda_1, \dots, \lambda_n$; the tables below are the paper's tables converted to 0-based
form.

```python
TABLES = {
    # Paper Figure 8 (converted to 0-based labels).
    "S_ex1": [[0, 1, 2], [1, 0, 1], [2, 1, 0]],
    "S_ex2": [[0, 1, 2], [1, 0, 1], [2, 1, 2]],
    # Paper Figure 5: the semigroup S_E^(2).
    "S_E2": [[0, 1, 2, 3], [1, 2, 3, 3], [2, 3, 3, 3], [3, 3, 3, 3]],
}
```

## Associativity check

`Semigroup.is_associative` verifies
$(\lambda_i \lambda_j) \lambda_k = \lambda_i (\lambda_j \lambda_k)$
for every triple of elements.

```python
for name, table in TABLES.items():
    semigroup = Semigroup(np.array(table))
    verdict = "is" if semigroup.is_associative else "is not"
    print(f"The semigroup {name} {verdict} associative")
```

## Result

Neither candidate table from Figure 8 is associative. For `S_ex1`, for instance,
$(\lambda_2 \lambda_2)\lambda_3 = \lambda_1 \lambda_3 = \lambda_3$ while
$\lambda_2(\lambda_2 \lambda_3) = \lambda_2 \lambda_2 = \lambda_1$ (1-based labels).
The table $S_E^{(2)}$ — a member of the family $S_E^{(N)}$ used throughout the S-expansion
literature — passes the test and is a genuine semigroup.
