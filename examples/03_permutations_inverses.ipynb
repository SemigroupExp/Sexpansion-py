# 3. Permutations of $n$ elements and their inverses

Isomorphism testing between semigroups (notebook 4) relies on permutations of the element labels:
two tables describe the same abstract semigroup exactly when some relabelling $\sigma$ maps one
table onto the other.

This notebook ports the Java example `I_Permutations_and_inverses_console_n4` from Section 5.4 of
the paper (C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). It enumerates all $N! = 24$ permutations of $N = 4$ symbols in the package's
canonical (lexicographic) order and prints each one together with its inverse.

The one-line notation $(a\ b\ c\ d)$ means the permutation maps $0 \mapsto a$,
$1 \mapsto b$, $2 \mapsto c$, $3 \mapsto d$ (0-based). The index `P#k` is the position in the
canonical enumeration, matching the labelling used by the isomorphism reports.

```python
from sexpansion import Permutation
```

## Parameters

```python
N = 4
```

## Enumerate the permutations

```python
permutations = list(Permutation.all_of_degree(N))
index = {p: i for i, p in enumerate(permutations)}
for i, p in enumerate(permutations):
    inverse = p.inverse()
    print(f"Permutation P#{i}: {p}")
    print(f"  inverse (P#{i})^-1 = P#{index[inverse]}: {inverse}")
```

## Result

Involutions — the identity `P#0`, the transpositions, and double transpositions such as
`P#16` — are their own inverses; the remaining permutations come in mutually inverse pairs such as
`P#9` and `P#18`.
