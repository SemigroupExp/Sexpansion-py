# 7. Resonances across the whole catalogue of one order

Notebook 6 found the resonant decompositions of two specific semigroups; here we scan the
**entire order-4 catalogue**. Only commutative semigroups are considered, since S-expansions
require abelian semigroups.

This notebook ports the Java example `II_findAllResonances_console_ord4` from Section 5.3 of the
paper (C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). Expected (paper, Table 5): **48** semigroups with at least one resonance and
**124** resonances in total.

```python
from sexpansion import find_all_resonances, load_semigroups
```

## Parameters

```python
ORDER = 4
```

## Scan the catalogue

```python
n_semigroups = 0
n_resonances = 0
for semigroup in load_semigroups(ORDER):
    if not semigroup.is_commutative:
        continue
    resonances = find_all_resonances(semigroup)
    if resonances:
        n_semigroups += 1
        n_resonances += len(resonances)
        print(f"The semigroup #{semigroup.sem_id} has {len(resonances)} resonances")

print(
    f"\nThere are {n_semigroups} semigroups with at least one resonance "
    f"and in total {n_resonances} different resonances."
)
```

## Result

48 of the 58 commutative order-4 semigroups admit at least one resonance, with 124 resonances in
total — reproducing Table 5 of the paper. The same scan runs for any catalogued order by changing
`ORDER` (orders 5 and 6 take substantially longer). Scans like this one are the starting point
when hunting for new S-expansions with prescribed properties.
