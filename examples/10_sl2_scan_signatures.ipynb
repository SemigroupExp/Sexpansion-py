# 10. Scanning all expansions of $\mathfrak{sl}(2)$ at a fixed order

Notebook 9 expanded $\mathfrak{sl}(2)$ with one hand-picked semigroup. Here we turn the
question around: for **every** commutative semigroup of a chosen order, compute the expanded
algebra $G_S$ and ask which expansions preserve semi-simplicity, classifying the survivors by the
eigenvalue signature of their Cartan–Killing metric.

This notebook ports the Java examples `III_sl2_SExp_ord2_to_6_console` and
`III_Signature_Sem_ord*` from Section 5.6 of the paper (C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). Expected for order 3 (paper,
Table 5): **12** commutative semigroups, of which **5** yield a semi-simple expanded algebra.

```python
from sexpansion import LieAlgebra, load_semigroups
```

## Parameters

The same scan runs at any catalogued order (2–6) by changing `ORDER`; orders 5 and 6 take
substantially longer.

```python
ORDER = 3
```

## Scan the catalogue

An expansion counts as semi-simple when $|\det g| > 10^{-6}$ — a safe threshold, since the
Cartan–Killing metric of an expanded algebra has integer entries.

```python
sl2 = LieAlgebra.sl2()

n_commutative = 0
n_semisimple = 0
for semigroup in load_semigroups(ORDER):
    if not semigroup.is_commutative:
        continue
    n_commutative += 1
    expanded = sl2.s_expand(semigroup)
    determinant = expanded.det()
    if abs(determinant) > 1e-6:
        n_semisimple += 1
        print(
            f"Semigroup #{semigroup.sem_id}: semi-simple expansion, "
            f"det = {determinant:.6g}, signature (+,-,0) = {expanded.signature()}"
        )

print(
    f"\nThere are {n_commutative} commutative semigroups of order {ORDER} "
    f"and {n_semisimple} expansions that give a semi-simple algebra."
)
```

## Result

Of the 12 commutative semigroups of order 3, exactly 5 give a semi-simple 9-generator expansion,
reproducing Table 5 of the paper. Four of them (semigroups #7, #10, #12, #16) have signature
$(6, 3, 0)$ — three times the $(2, 1, 0)$ signature of $\mathfrak{sl}(2, \mathbb{R})$ itself —
while #18 gives $(5, 4, 0)$, a genuinely different real form. Scans like this one, combined with
the resonance scans of notebook 7, are how new physically interesting algebras are found with the
S-expansion method.
