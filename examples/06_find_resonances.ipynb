# 6. Resonant decompositions of semigroups from the literature

A **resonance** of a semigroup is a decomposition $S = S_0 \cup S_1$ satisfying

$$S_0 \cdot S_0 \subset S_0, \qquad S_0 \cdot S_1 \subset S_1, \qquad
S_1 \cdot S_1 \subset S_0,$$

mirroring the graded structure $[V_0, V_0] \subset V_0$, $[V_0, V_1] \subset V_1$,
$[V_1, V_1] \subset V_0$ of a Lie algebra split into a subalgebra $V_0$ and a complement $V_1$.
A resonant decomposition is what allows a **resonant subalgebra** to be extracted from the
expanded algebra (notebook 9).

This notebook ports the Java example `II_findresonances_ex_console` from Section 5.2 of the paper
(C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). The semigroups $S_E^{(2)}$ and $S_{N3}$ (paper, Figures 5 and 14) were used in the
literature to obtain the Bianchi algebras as S-expansions; here we compute all of their resonant
decompositions.

```python
import numpy as np

from sexpansion import Semigroup, find_all_resonances
```

## The multiplication tables

From paper Figures 5 and 14, converted to 0-based labels.

```python
SEMIGROUPS = {
    "S_E2": [[0, 1, 2, 3], [1, 2, 3, 3], [2, 3, 3, 3], [3, 3, 3, 3]],
    "S_N3": [[3, 0, 3, 3], [0, 1, 2, 3], [3, 2, 3, 3], [3, 3, 3, 3]],
}
```

## Find all resonances

`find_all_resonances` enumerates every resonant decomposition; the subsets $S_0$ and $S_1$ are
printed as tuples of 0-based element indices.

```python
for name, table in SEMIGROUPS.items():
    resonances = find_all_resonances(Semigroup(np.array(table)))
    print(f"The semigroup {name} has {len(resonances)} resonances:")
    for k, resonance in enumerate(resonances, start=1):
        print(f"  Resonance #{k}: S0 = {resonance.s0}, S1 = {resonance.s1}")
```

## Result

$S_E^{(2)}$ admits 2 resonant decompositions and $S_{N3}$ admits 5. Both semigroups have a zero
element ($\lambda_4$, index 3), and it always appears in **both** subsets: since
$0_S \cdot \lambda = 0_S$ for every $\lambda$, the closure conditions
$S_0 \cdot S_1 \subset S_1$ and $S_1 \cdot S_1 \subset S_0$ force the zero into $S_0$ and
$S_1$ alike.
