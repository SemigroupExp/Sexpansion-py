# 4. Finding the catalogued semigroup isomorphic to a given table

Two semigroups are isomorphic when a permutation of the element labels carries one multiplication
table onto the other. The bundled catalogue contains exactly one representative of each
isomorphism class, so any hand-entered table is isomorphic to exactly one catalogued semigroup.

This notebook ports the Java example `I_isomorphisms_ex2` from Section 5.4 of the paper (C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)).
We take the semigroup $S_{N3}$ (paper, Figure 14), search the order-4 catalogue for its
representative, and then list every permutation relating the two tables.

Expected (paper, Table 4): the representative is $S^{42}_{(4)}$, related by the permutation
`P#19` — written $(4\ 1\ 3\ 2)$ in the paper's 1-based convention, i.e. $(3\ 0\ 2\ 1)$ in
the 0-based convention used by the code.

```python
import numpy as np

from sexpansion import Permutation, Semigroup, load_semigroups
```

## The semigroup $S_{N3}$

The table from paper Figure 14, converted to 0-based labels.

```python
SN3 = Semigroup(np.array([[3, 0, 3, 3], [0, 1, 2, 3], [3, 2, 3, 3], [3, 3, 3, 3]]))
```

## Search the order-4 catalogue

`isomorphism_test` decides whether two semigroups are isomorphic; once the representative is
found, a brute-force scan over the 24 permutations of degree 4 lists every relabelling
$\sigma$ with $\sigma(S^{42}_{(4)}) = S_{N3}$.

```python
for candidate in load_semigroups(4):
    if candidate.isomorphism_test(SN3).isomorphic:
        print(f"The semigroup #{candidate.sem_id}")
        print(candidate)
        print("is isomorphic to SN3.")
        for k, sigma in enumerate(Permutation.all_of_degree(4)):
            if candidate.permute(sigma) == SN3:
                print(f"A permutation that brings #{candidate.sem_id} to SN3 is P#{k}: {sigma}")
                print(f"  its inverse is {sigma.inverse()}")
```

## Result

Catalogue entry $S^{42}_{(4)}$ is the representative of $S_{N3}$, as stated in Table 4 of the
paper. Two permutations relate the tables, `P#19` and `P#22`: an isomorphism need not be unique,
since composing one relabelling with any automorphism of the table yields another.
