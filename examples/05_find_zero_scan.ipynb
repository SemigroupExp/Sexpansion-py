# 5. Commutative semigroups with a zero element

A **zero element** $0_S$ of a semigroup satisfies $0_S \cdot \lambda_i = 0_S$ for every element
$\lambda_i$. Zero elements are central to the S-expansion method: when $S$ has a zero, the
expanded algebra $G_S = S \otimes G$ admits a $0_S$-**reduction** that discards the generators
multiplied into the zero, producing a smaller algebra (see notebooks 9 and 10).

This notebook ports the Java example `II_findzero_console_ord4` from Section 5.2 of the paper
(C. Inostroza, I. Kondrashuk, N. Merino, F. Nadal, *A Java Library to Perform S-Expansions of Lie Algebras*, Axioms (2025)). It scans the order-4 catalogue for commutative semigroups possessing a zero element and
prints each one. Expected (paper, Table 5): **39** of the 58 commutative semigroups of order 4
have a zero.

```python
from sexpansion import load_semigroups
from sexpansion.reports import semigroup_report
```

## Parameters

```python
ORDER = 4
```

## Scan the catalogue

The reports are printed with `one_based=True`, matching the paper's
$\lambda_1, \dots, \lambda_n$ labelling.

```python
count = 0
for semigroup in load_semigroups(ORDER):
    zero = semigroup.find_zero()
    if semigroup.is_commutative and zero is not None:
        count += 1
        print(f"#{semigroup.sem_id}")
        print(semigroup_report(semigroup, one_based=True))
        print(f"The zero element is {zero + 1} (1-based label)\n")
print(f"Number of semigroups with zero element: {count}")
```

## Result

39 commutative order-4 semigroups have a zero element, reproducing Table 5 of the paper. Each of
them is a candidate for building $0_S$-reduced S-expansions.
